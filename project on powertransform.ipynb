```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("C:\\Users\\subha\\OneDrive\\Documents\\concrete_data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x=df.drop(columns=["concrete_compressive_strength"])
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ir=LinearRegression()
ir.fit(x_train,y_train)

y_pred=ir.predict(x_test)

r2_score(y_test,y_pred)
```

```python
ir=LinearRegression()
np.mean(cross_val_score(ir,x,y,scoring="r2"))
```

```python
for col in x_train.columns:
    
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(x_train[col],dist="norm",plot=plt)
    plt.title(col)
    plt.show()
```

```python
pt_bc=PowerTransformer(method="box-cox")
x_train_transformed_bc=pt_bc.fit_transform(x_train+0.000001)
x_test_transformed_bc=pt_bc.transform(x_test+0.000001)
# p=pd.DataFrame({"cols":x_train.columns,"box_cox_lambdas":pt_bc.lambdas_})
```

```python
ir=LinearRegression()
ir.fit(x_train_transformed_bc,y_train)

y_pred2=ir.predict(x_test_transformed_bc)

r2_score(y_test,y_pred2)
```

```python
pt_bc=PowerTransformer(method="box-cox")
x_transformed_bc=pt_bc.fit_transform(x+0.000001)

ir=LinearRegression()
np.mean(cross_val_score(ir,x_transformed_bc,y,scoring="r2"))
```

```python
x_train_transformed_bc=pd.DataFrame(x_train_transformed_bc,columns=x_train.columns)
for col in x_train_transformed_bc.columns:
    
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)

    plt.subplot(122)
    sns.distplot(x_train_transformed_bc[col])
    plt.title(col)
    plt.show()
```

```python
pt_yj=PowerTransformer(method="yeo-johnson")
x_train_transformed_yj=pt_yj.fit_transform(x_train+0.000001)
x_test_transformed_yj=pt_yj.transform(x_test+0.000001)
# q=pd.DataFrame({"cols":x_train.columns,"yeo_johnson_lambdas":pt_yj.lambdas_})
```

```python
ir=LinearRegression()
ir.fit(x_train_transformed_yj,y_train)

y_pred3=ir.predict(x_test_transformed_yj)

r2_score(y_test,y_pred3)
```

```python
pt_yj=PowerTransformer(method="yeo-johnson")
x_transformed_yj=pt_yj.fit_transform(x+0.000001)

ir=LinearRegression()
np.mean(cross_val_score(ir,x_transformed_yj,y,scoring="r2"))
```

```python
x_train_transformed_yj=pd.DataFrame(x_train_transformed_bc,columns=x_train.columns)
```

```python
for col in x_train_transformed_yj.columns:
    
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)

    plt.subplot(122)
    sns.distplot(x_train_transformed_yj[col])
    plt.title(col)
    plt.show()
```

```python
model_bc=LinearRegression()
model_bc.fit(x_train_transformed_bc,y_train)
```

```python
model_yj=LinearRegression()
model_yj.fit(x_train_transformed_bc,y_train)
```

```python
score_bc=model_bc.score(x_test_transformed_bc,y_test)
score_yj=model_yj.score(x_test_transformed_yj,y_test)
```

```python
print(f"score_bc:{score_bc:.3f}")
print(f"score_yj:{score_yj:.3f}")
```
