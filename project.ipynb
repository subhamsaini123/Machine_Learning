# custom feature engineering

```python
x=np.array([[1,2],[3,4]])
x
```

```python
def function_engineering(x):
    return np.hstack((x,x**2))
```

```python
custom_transform=FunctionTransformer(function_engineering)
```

```python
x_transformed=custom_transform.transform(x)
x_transformed
```

# scaling and normalization

```python
x=np.array([[1,2],[3,4]])
x
```

```python
def scaling(x):
    return x/np.max(x)
```

```python
custom_transform=FunctionTransformer(scaling)
```

```python
x_transformed=custom_transform.transform(x)
```

```python
x_transformed
```
