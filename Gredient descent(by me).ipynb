```python
class RegexKaGD:
    def __init__(self,learning_rate,epochs):
        self.m=25.49
        self.b=0
        self.lr=learning_rate
        self.epochs=epochs
    def fit(self,X,y):
        for i in range(self.epochs):
            loss_slope=-2*np.sum(y-self.m*X.ravel()-self.b)
            step_size=self.lr*loss_slope
            self.b=self.b-(step_size)
        print(self.b)
```
